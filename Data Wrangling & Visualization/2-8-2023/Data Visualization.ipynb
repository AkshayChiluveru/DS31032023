```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the tips data from seaborn library

### Note : tips dataset is in built dataset available in the seaborn library. When you write the below code it will load the data directly from the seaborn library

```python
data = sns.load_dataset('tips')
data.head()
```

## Compare whether Male are paying the highest bill or Female are paying the highest bill

```python
sns.barplot(x = data['sex'], y = data['total_bill'])
### Total bill Comparison
plt.title("Total Bill Comparison", fontsize = 16)
plt.ylabel('Total Bill', fontsize = 14)
plt.xlabel('Gender', fontsize = 14)
plt.show()
```

```python
sns.barplot(x = data['sex'], y = data['total_bill'], hue = data['smoker'])
plt.show()
```

## Count the Male and Female smoker


```python
sns.countplot(x = data['smoker'], hue = data['sex'])
plt.show()
```

## Observations:

- Male smoker are twice as female smokers
- Male Non - smoker are twice as female non smokers

## Cat plot

```python
### Compare whether Male are paying the highest bill or Female are paying the highest bill

sns.catplot(x = 'sex', y = 'total_bill', data = data, kind = 'bar', hue = 'smoker')
plt.show()
```

```python
## Count of Male and Female smokers

sns.catplot(x = 'smoker', data = data, kind = 'count', hue = 'sex')
```

## Bivariate Analysis - Two Variable Analysis (Two columns analysis)

- if both columns numerical - scatterplot, jointplot, regplot, etc
- if one column object another columns numerical - Bar plot, swarm plot, violin plot
- if one column is date another column numerical - Line plot

```python
data.head()
```

```python
sns.scatterplot(x = data['total_bill'], y = data['tip'])
```

```python
sns.regplot(x = data['total_bill'], y = data['tip'])
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = data)
```

```python
sns.heatmap(data.corr(), annot = True)
```

```python
sns.pairplot(data = data)
```

### Correlation plot or heatmap() is used to check the relation between numerical columns of the data 

### pairplot() is used to display scatterplot(if the columns are different) and histogram(if the columns are same) for all numerical columns present in the data
