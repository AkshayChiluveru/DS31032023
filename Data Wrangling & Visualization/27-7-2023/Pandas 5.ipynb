### Create Age bracket from the Age column where the conditions are 
1. 0 - 20 = Young
2. 20 - 40 = Adult
3. 40 - 60 = Old
4. '>' 60 = Retired

```python
def age_bracket(x):
    if x < 20:
        return 'Young'
    elif x >= 20 and x < 40:
        return 'Adult'
    elif x >= 40 and x < 60:
        return 'Old'
    else:
        return 'Retired'
```

```python
data['Age Bracket'] = data['Age'].apply(age_bracket)
data
```
