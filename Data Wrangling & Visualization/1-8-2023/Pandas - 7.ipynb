## Load the BigMart Sales data

```python
df = pd.read_csv('BigMart Sales.csv')
df.head()
```

## Which Outlet_Location_type is leading in total sales?

```python
#### Display the max sales for each Outlet location type

df.groupby('Outlet_Location_Type')['Item_Outlet_Sales'].agg('max')
```

```python
## Graphical method

sns.barplot(x = df['Outlet_Location_Type'], y = df['Item_Outlet_Sales'])
plt.show()
```

### When you are "comapring" the categories against some numerical data 
### then the choice is bar plot

```python
df['Outlet_Size'].unique()
```

```python
sns.barplot(x = df['Item_Outlet_Sales'], y = df['Outlet_Size'])
plt.show()
```
