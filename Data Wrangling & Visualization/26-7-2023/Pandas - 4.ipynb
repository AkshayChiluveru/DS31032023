### Display the maximum World Population Percentage for each Continent

- For each, For every - groupby()

```python
data.groupby('Continent')['World Population Percentage'].agg('max')
```

### Display the max and min value of World Population Percentage for every Continent

```python
data.groupby('Continent')['World Population Percentage'].agg(['max', 'min'])
```

### Always use agg() with groupby() otherwise without agg() the groupby output is meaningless

```python
data.groupby('Continent').agg('max')
```
