## Data Wrangling : 

- The process of treating missing values and combining complex datasets to make them more accessible and easier to analyze.
- To perform the above process of analysing data we use some of the standard libraries available in python namely,
1. pandas - Data Analysis & Data Wrangling
2. numpy - Data Analysis & Data Wrangling
3. matplotlib - Data Visualization
4. seaborn - Data Visualization
5. sklearn - Data Wrangling but the mainly used for Machine Learning

#### What is numpy?

- It is a python library, abbreviated as Numerical Python (NumPy)
- deals with n dimensional array data structure
- It is mainly used for $numerical$ analysis on data

### What is pandas

- Pandas is a python library, abbreviated as Panel Data (Pandas)
- deals with one dimensional series data structure and 2D DataFrames
- It is mainly used for Analysis of data
