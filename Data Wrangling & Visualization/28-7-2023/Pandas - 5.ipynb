## Loading the Standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('zomato.csv', encoding = 'latin1')
data.head()
```

## checking the columns available in the data

```python
data.columns
```

## Total rows and total columns in the data

```python
data.shape
```

- Observation :

### Delete unnecessary columns from the data

```python
cols_to_del = ['Country Code', 'Address', 'Locality', 'Locality Verbose', 'Longitude', 'Latitude',
              'Switch to order menu', 'Rating color']
cols_to_del
```

```python
data = data.drop(cols_to_del, axis = 1)
data.head()
```

```python
data.columns
```

### Dropping Duplicate rows from the data

```python
data = data.drop_duplicates()
data.head()
```

```python
data.shape
```

### Proportion of Excellent and Very Good rated restaurants

```python
data['Rating text'].value_counts(normalize = True) * 100
```

### Observations:
1. Lot of Resaurants have been rated as average(39.12% )
2. Only 3.15% restaurants are Excellent rated 
3. 11.29 % of restaurants rated as Very Good

### Total count of restaurants having aggregate rating greater than or equal to 4

```python
data[data['Aggregate rating'] >= 4]['Restaurant Name'].count()
```

```python
1380 / 9551 * 100
```

#### Observations :
- Out of 9551 restaurants only 1380 have aggregate ratings >= 4 (14.44%)
